```python
import requests
from bs4 import BeautifulSoup
```

```python
# file_path = '../cache/2109.11017.pdf'
file_path = '../cache/1810.12346v1.pdf'
```

```python
TIKA_URL = 'http://localhost:9998/tika/form'
```

```python
files=[
  ('file',(file_path.split('/')[-1],open(file_path,'rb'),'application/pdf'))
]
headers = {
  # 'Accept': 'text/plain'
}
```

```python
response = requests.post(TIKA_URL, files=files, headers=headers)
```

```python
response.raise_for_status()
```

```python
soup = BeautifulSoup(response.content.decode('utf-8'), 'lxml')
```
